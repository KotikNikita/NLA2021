## Task 1
#### 1.1  
We have to find $C_1$ and $C_2$ such that $\forall x $: $C_1 \sum\limits_{i=1}^n |x_i| \leq \max\limits_{i \in \{1...n\}}|{x_i}| \leq C_2 \sum\limits_{i=1}^n |x_i|$. Let's define $x_{max} = \max\limits_{i \in \{1...n\}}|{x_i}|$.  
-  $C_1$: Since $\forall i: |x_i| \leq x_{max} \Rightarrow$ $\sum\limits_{i=1}^n |x_i| \leq n \cdot x_{max} \Rightarrow$
$\frac{1}{n} \sum\limits_{i=1}^n |x_i| \leq \max\limits_{i \in \{1...n\}}|{x_i}| \Rightarrow$ $C_1 = \frac{1}{n}$  
-  $C_2$: It follows from the definition of $x_{max}$ that $\exists i \in \{1...n\}: |x_i| = x_{max} \Rightarrow$ $\max\limits_{i \in \{1...n\}}|{x_i}| \leq \sum\limits_{i=1}^n |x_i| \Rightarrow$ $C_2 = 1$  

#### 1.2  
We have to prove that $\| U A \|_F = \| A U \|_F = \| A \|_F$ for any unitary matrix $U$.  
Complex $n\times n$ square matrix is called **unitary** if  
$$ U^*U = UU^* = I_n, $$  
We know that $\| A \|_F = \sqrt{\left(A,A\right)_F}$, where $\left(A,B\right)_F = trace(A^*B) \Rightarrow$  
$\| U A \|_F^2 = trace((UA)^*UA) = trace((A^*U^*UA) = trace((A^*A) = \| A \|_F^2 \Rightarrow$ $\| U A \|_F = \| A \|_F$  
Similarly: $\| A U \|_F^2 = trace((AU)^*AU) = trace(U^*A^*AU) = trace(UU^*A^*A) = trace(A^*A) = \| A \|_F^2 \Rightarrow$ $\| A U\|_F = \| A \|_F$  
#### 1.3  
We have to prove that $\| U A \|_2 = \| A U \|_2 = \| A \|_2$ for any unitary matrix $U$.  
Spectral norm:  
$$ \Vert A \Vert_2 = \sigma_1(A) = \sqrt{\lambda_\max(A^*A)} = \sup\limits_{\|z\|_2=1}\|Az\|_2 $$  
Also we know that complex square matrix $U$ is unitary iff $\Vert U z \Vert_2 = \Vert z \Vert_2$ for all $z$.  
$\| U A \|_2 = \sup\limits_{\|z\|_2=1}\|U\underbrace{Az}_{vector}\|_2 = \{\textit{since U is unitary}\}= \sup\limits_{\|z\|_2=1}\|Az\|_2 = \| A \|_2$.  
Similarly: $\| A U \|_2 = \sup\limits_{\|z\|_2=1}\|AUz\|_2 = \{\textit{since U is unitary}\} = \sup\limits_{\|Uz\|_2=1}\|A\underbrace{Uz}_{\textit{vector }x}\|_2 = \sup\limits_{\|x\|_2=1}\|Ax\|_2 = \| A \|_2$
 
## Task 2
#### 2.1  
We have to prove that $\| A \|_F \le \sqrt{\mathrm{rank}(A)} \| A \|_2$. We will use following  
Theorem. Any matrix $A\in \mathbb{C}^{n\times m}$ can be written as a product of three matrices:
$$ A = U \Sigma V^*, $$
where

$U$ is an $n \times n$ unitary matrix  
$V$ is an $m \times m$ unitary matrix  
$\Sigma$ is a diagonal matrix with non-negative elements $\sigma_1 \geq \ldots, \geq \sigma_{\min (m,n)}$ on the diagonal.

Moreover, if $\text{rank}(A) = r$, then $\sigma_{r+1} = \dots = \sigma_{\min (m,n)} = 0$  

Let's do it:  
$\|A\|_F^2 = \{\textit{SVD decomposition}\} = \|U \Sigma V^*\|_F^2 = \{\textit{follows from task 1.2}\} = \|\Sigma\|_F^2 = \{\textit{definition of Frobenius norm}\} = \sum\limits_{i=1}^{min(n,m)}\sigma^2_i \leq$  
$\{\textit{prorerty of  }\Sigma \} \leq \sigma^2_1 \cdot \left( \sum\limits_{i=1}^{min(n,m)}I[\sigma_i>0] \right) = \{\textit{follows from theorem above}\} = \mathrm{rank}(A) \cdot \sigma_1^2 = \mathrm{rank}(A) \cdot \| A \|^2_2$  
After taking the square root, we get what we need.

#### 2.2  
We have to show that for any $n, m$ and $k \le \min(n, m)$ there exists $A \in \mathbb{R}^{n \times m}: \mathrm{rank}(A) = k$, such that $\| A \|_F = \sqrt{\mathrm{rank}(A)} \| A \|_2$. In other words, that the previous inequality is not strict.  
Such a matrix is easy to construct simply by looking at the proof of the previous subtask.Matrix $A$ will immediately have the form of $\Sigma$ matrix from a singular decomposition (in other words, $U$ and $V$ will be $I_n$ and $I_m$ respectively). We just have to choose the right singular values: first $k$ of them must be equal nonzero, others are zeroes. So, for given $n$,$m$,$k$:
$$ (A)_{ij} = \begin{cases} 1 & \textit{ if }i=j \textit{ and } i \in \{1...k\} \\ 0 & \text{otherwise}, \end{cases} $$  
It's easy to see, that for such a matrix inequalities becomes equalities in chain from previous subtask. Or that $\| A \|_F = \sqrt{k} = \sqrt{\mathrm{rank}(A)}$ and $\| A \|_2=1 \Rightarrow$ $\| A \|_F = \sqrt{\mathrm{rank}(A)} \| A \|_2$

#### 2.3  
We have to prove that if $A \in \mathbb{R}^{n \times n}$ - diagonalizable matrix, then $\det(\exp(A)) = \exp(\mathrm{trace}(A))$.  
$A \in \mathbb{R}^{n \times n}$ - **diagonalizable** matrix, if
$$ \exists \text{ invertible matrix } S \in \mathbb{R}^{n \times n} \text{ and diagonal matrix }D\in \mathbb{R}^{n \times n}: A = S^{-1}DS $$  
  
We will not prove hits, since their proof is trivial. Just use them:  
  
$\det(\exp(A)) = \det(\exp(S^{-1}DS)) = \{\text{hint}\} = \det(S^{-1}\exp(D)S) = \det(S^{-1})\det(\exp(D))\det(S) = \det(\exp(D)) = \{\text{hint}\} = \det(\mathrm{diag}(\exp(d_1), \exp(d_2), ..., \exp(d_n))) = \exp(d_1) \cdot \exp(d_2) \cdot ... \cdot\exp(d_n) = \exp\left({\sum\limits_{i=1}^nd_i}\right) = \exp(\mathrm{trace}(D)) =$  
=$ \exp(\mathrm{trace}(SS^{-1}D)) = \exp(\mathrm{trace}(S^{-1}DS)) = \exp(\mathrm{trace}(A))$.

#### 2.4  
We have to prove that $\| A B \|_2 \le \| A \|_2 \| B \|_F$.  
First remark: $\|A\|_2 = \sup\limits_{\|x\| \not = 0}\frac{\|Ax\|_2}{\|x\|_2} \geq \frac{\|Ax\|_2}{\|x\|_2} \Rightarrow \|Ax\|_2 \leq \|A\|_2 \|x\|_2 \text{ } \forall x$.  
Second remark: $\|A\|_F = \sqrt{\sum\limits_{i=1}^{min(n,m)} \sigma_i^2} \geq \sigma_1 = \|A\|_2$, more details in 2.1  
Finally: $\|AB\|_2 = \sup\limits_{\|x\| = 1}\|A\underbrace{Bx}_{vector}\|_2 \leq \{\text{first remark}\} \leq \sup\limits_{\|x\| = 1}\|A\|_2\|Bx\|_2 = \|A\|_2 \cdot \|B\|_2 \leq \{\text{second remark}\} \leq \|A\|_2 \|B\|_F$

### Second step: neural network architecture

For simplicity and demonstration purposes of the neural network compression idea consider the architecture consisting of the only fully-connected layers and non-linear ReLU functions between them.
To demonstrate compression effect, consider the dimension of the inner layers equals to 1000.

Below you see implementation of such neural network in PyTorch.
More details about neural networks you will study in the *Deep learning* course in one of the upcoming term

````python
class ResidualBlock(nn.Module):
    def __init__(self, in_features, middle_features, activation=nn.ReLU()):
        super().__init__()
        self.layer1 = nn.Linear(in_features, middle_features)
        self.layer2 = nn.Linear(middle_features, in_features)
        self.activation = activation
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.activation(out)
        out = self.layer2(out)
        out = out + x
        out = self.activation(out)
        return out

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3 * 32 * 32, 1000)
        self.res1 = ResidualBlock(1000,1000)
        self.res2 = ResidualBlock(1000,1000)
        self.fc2 = nn.Linear(1000, 10)
        self.ReLU = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.res1(x)
        x = self.res2(x)
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)
````

#### Implement functions for training and testing after every sweep over all dataset entries

````python
def train(model, train_loader, optimizer, epoch):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.train()
    model.to(device)
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
````

````python
def test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    device = "cuda" if torch.cuda.is_available() else "cpu"
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
    return correct
````
